```python
import pandas as pd
```

```python
data=pd.read_csv('ml2_train.csv')
```

```python
data.isnull().sum()
```

```python
data.fillna(0,inplace=True)
```

```python
data.head()
```

```python
t10=data['2010 Deposits'].tolist()
t11=data['2011 Deposits'].tolist()
t12=data['2012 Deposits'].tolist()
t13=data['2013 Deposits'].tolist()
t14=data['2014 Deposits'].tolist()
t15=data['2015 Deposits'].tolist()
```

```python
lis=[]
res=[]
check=[]
for i in range(0,int(len(data))):
    lis.append([t10[i],t11[i],t12[i],t13[i],t14[i]])
    res.append([t15[i]])
    check.append([t11[i],t12[i],t13[i],t14[i],t15[i]])
```

```python
from sklearn import tree
#features = [[140,1], [130,1], [150,0], [170, 0]]
#labels = [0,0,1,1]

clf = tree.DecisionTreeClassifier()
clf = clf.fit(lis, res)
a= (clf.predict(check))
```

```python
result=pd.Series((a.tolist()))
result
```

```python
new=pd.DataFrame(result,columns=['2016 Deposits'])
new['Serial Number']=range(0,int(len(data)))
new.drop('2016 Deposits',axis=1,inplace=True)
new['Deposits 2016']=result
new=new.sort_values('Deposits 2016')
new=new.reset_index()
new.drop('index',1,inplace=True)
new.to_csv('Results.csv',index=False)
```

```python
new[new[new['Deposits 2016']>=20000]<=90000].sort_values('Deposits 2016').dropna()
```

```python
first=new[new[new['Deposits 2016']>=20000]<=90000].sort_values('Deposits 2016').dropna()
sec=new[new['Deposits 2016']<20000].sort_values('Deposits 2016').dropna()
third=new[new['Deposits 2016']>20000].sort_values('Deposits 2016').dropna()
```

```python
final=first.append(sec).append(third).round(0)
```

```python
final['Serial Number']=final['Serial Number'].astype(int)
final['Deposits 2016']=final['Deposits 2016'].astype(int)
```

```python
final.to_csv('Results.csv',index=False)
```
